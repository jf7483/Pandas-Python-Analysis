## Broadcasting Operations 

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba["Age"].add(5)
# add five to every value within the series 
nba["Age"] + 5 
# subtracting 5 million from each row 
nba["Salary"].sub(5000000)
nba["Salary"] - 5000000

# .mul is the multiply method that we want to multiply each row with 
nba["Weight"].mul(0.453592)
nba["Weight"] * 0.45392
```

```python
# lets say that you want to save to kg you calculated above to your dataframe 
nba["Weight in Kg"] = nba["Weight"] * 0.453592
```

```python
# weight in kg has been added to the right of the dataframe 
nba.head(3)
```

```python
nba["Salary"].div(1000000)
# the things that are on the right happen first 
nba["Salary in Millions"] = nba["Salary"].div(1000000)
```

```python
nba.head(3)
```

## Drop Rows with NULL Values

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.tail(3)
```

```python
# dropna removes any rows in the dataframe that have NULL values 
# the parameter how = any says to remove ANY rows that have a NULL value 
nba.dropna()
```

```python
# this means to drop the rows where ALL the values are NULL and as we can see some of the 
# players that were removed using the syntax above come back because it has other 
# values that are not null. The inplace permanently removes them 
nba.dropna(how = "all", inplace = True)
```

```python
nba.tail(3)
```

```python
# is we want to drop any columns that have null values we do the following 
#for the axis parameter in the method, we will set it equal to 1 because that denotes the 
# column whereas 0 denotes the rows 
# we can see that college and salary have been removes but it is not permanent unless we set
# the inplace parameter. It removes those because of the how = any 
nba.dropna(axis = 1)
```

```python
# if we want to remove a specific row if there is a null in the salary column 
# but we don't care if there is a null in any other column 

# this will drop the rows wherever there is a N`ULL value in the salary column and college column
nba.dropna(subset = ['Salary', 'College'])
```

## The .astype() Method 

```python
# drops all the rows where the value is null 
nba = pd.read_csv("nba.csv").dropna(how='all')
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.tail()
```

```python
nba.dtypes
nba.info()
```

```python
# we are converting the float type in the salary to a int type 
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.info()
```

```python
nba["Age"] = nba["Age"].astype("int")
nba.info()
```

```python
nba["Number"] = nba["Number"].astype("int")
nba.head(3)
```

```python
# we have 5 unique positions in NBA
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba.info()
```

```python
# memory usage will go further down 
nba["Team"].nunique()
nba["Team"] = nba["Team"].astype("category")
```

## Sort DataFrame with the .sort_values() Method Part 1 

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# the first parameter is the string of the column that you want to sort by
# if you use the Name column it will use that column to sort all the rows 
nba.sort_values("Name", ascending = False)
```

```python
# show the oldest age first 
nba.sort_values("Age", ascending = False)
```

```python
# the greatest salary is first and preview the first three
# inplace sets it so that when we preview it it is what we see with the bottom one 
nba.sort_values("Salary", ascending = False, inplace = True)
```

```python
nba.head(3)
```

```python
# this will usually not make sense but if we want the NULL values first this is how we would do it 
nba.sort_values("Salary", ascending = False, na_position = 'first')

# to preview the last few rows with the smallest 
nba.sort_values("Salary", ascending = False, na_position = 'first').tail()
```

## Sort a DataFrame with the .sort_values() Method Part II 

### this will show you how to sort with multiple columns

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# when we want to specify more than one thins, we would just want to specify a python list 

# we can see that it first sorts by the Team and then by the Name. It sorts them by alphabetical order 
nba.sort_values(["Team", "Name"])
```

```python
# both columns in reverse alphabetical order 
nba.sort_values(["Team", "Name"], ascending = False)
```

```python
# what if we want to sort the team column in asc order and sort the names in desc (reverse alpha)
# we can provide a list of booleans 

# team will be sorted alphabetically and then the name will be in descending order
# inplace to make it valid 
nba.sort_values(["Team", "Name"], ascending = [True, False], inplace = True)
```

```python
nba.head(3)
```

## Sort DataFrame with the .sort_index() Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Number", "Salary","Name"], inplace = True)
nba.tail(3)
```

```python
# sort by index so that we can get back to the original shape 

# by default it will be sorted by ascneding order 
nba.sort_index()

# to sort by the descending index and becase of the na_position it will put the NaN values at the end 
nba.sort_index(ascending = False, inplace = True)
```

```python
# will show descending (greatest to lowest) 
nba.head(3)
```

## Rank Values with the .rank() Method 

### call on a series to generate a new series of ranks

```python
# to get the rank method to work we will need to get rid of the NULL values first 
# the how = 'all' means that it will get rid of the rows that have all NULL values so it knocks out the 
# the last row 
nba = pd.read_csv("nba.csv").dropna(how = "all")

# next we want to fill in the null values in the salary column with a 0 and convert the float to int 
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
# what it is doing is ranking the highest salary towards the bottom 
# we can make a new column that displays the ranks for each salary 

nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
```

```python
nba.head(3)
```

```python
nba.sort_values(by = "Salary Rank")
```
