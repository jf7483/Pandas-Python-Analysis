```python
import pandas as pd
import numpy as np 
```

## Changing Options with Attributes and Dot Syntax 

```python
# telling numpy it wants to generate random integers from 0 to 100 
# the next argument is tha size of the array so 1000 rows and 50 columns 

data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.tail(2)
```

```python
pd.options.display.max_rows = 18
```

```python
pd.options.display.max_columns = 8
```

```python
df
```

## Changing Pandas Options with methods

```python
data = np.random.randint(0, 100, [1000, 50])
df = pd.DataFrame(data)
df.tail(2)
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 16)
```

```python
df
```

```python
pd.set_option("max_columns", 20)
```

```python
df
```

```python
pd.options.display.max_columns
```

```python
pd.reset_option("max_rows")
```

```python
pd.get_option("max_rows")
```

```python
pd.describe_option('max_columns')
```
